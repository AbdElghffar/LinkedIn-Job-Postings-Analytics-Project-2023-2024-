```python
# Define columns that are irrelevant, duplicated, or not needed for analysis
cols_to_drop = [
'company_name',
'max_salary', 'min_salary', 'med_salary',
'work_type',
'currency', 'compensation_type',
'job_posting_url', 'application_url',
'original_listed_time', 'listed_time', 'expiry', 'closed_time',
'zip_code', 'fips'
]
```

```python
# Drop the unnecessary columns from the dataset
postings_clean = postings.drop(columns=cols_to_drop)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
postings_clean = postings_clean.drop(columns=[
'skills_desc',
'remote_allowed',
'pay_period'
])
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Remove rows where company_id is missing (important for linking with companies data)
postings_clean = postings_clean.dropna(subset=['company_id'])
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Check average salary by job title
postings_clean.groupby(['title', 'location'])['normalized_salary'].mean().sort_values(ascending=False).head(10)
```

```python
postings_clean['normalized_salary'] = postings_clean.groupby(['title', 'location'])['normalized_salary'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Fill missing salary using average salary per title
postings_clean['normalized_salary'] = postings_clean.groupby('title')['normalized_salary'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Fill applies based on similar jobs (title)
postings_clean['applies'] = postings_clean.groupby('title')['applies'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Use views to estimate applies (ratio)
ratio = postings_clean['applies'].sum() / postings_clean['views'].sum()

postings_clean['applies'] = postings_clean['applies'].fillna(
    postings_clean['views'] * ratio
)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Fill based on title
postings_clean['formatted_experience_level'] = postings_clean.groupby('title')[
    'formatted_experience_level'
].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Fill remaining missing values with overall median
postings_clean['normalized_salary'] = postings_clean['normalized_salary'].fillna(
    postings_clean['normalized_salary'].median()
)
```

```python
# Fill missing views using average views per job title
postings_clean['views'] = postings_clean.groupby('title')['views'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Fill posting_domain using the most frequent domain per company
postings_clean['posting_domain'] = postings_clean.groupby('company_id')['posting_domain'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
)
```

```python
# Fill remaining values using application_type groups
postings_clean['posting_domain'] = postings_clean.groupby('application_type')['posting_domain'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
)
```

```python
# Drop rows with any remaining missing values
postings_clean = postings_clean.dropna()
```

```python
postings_clean.isnull().sum().sort_values(ascending=False) / len(postings_clean) * 100 
```

```python
# Save cleaned postings dataset to CSV
postings_clean.to_csv(r"E:/Graduation Project(DEPI)/clean data/postings_clean.csv", index=False)
```
