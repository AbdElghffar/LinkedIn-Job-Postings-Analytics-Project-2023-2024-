```python
def make_annual(row): # Making Annual Salary
    salary = row['avg_salary']
    period = row['pay_period']
    

    if pd.isna(salary):
        return np.nan
        

    if period == 'YEARLY':
        return salary
    elif period == 'HOURLY':
        return salary * 2080
    elif period == 'MONTHLY':
        return salary * 12
    elif period == 'WEEKLY':
        return salary * 52
    elif period == 'BIWEEKLY':
        return salary * 26
    else:
        return salary

jobs_df_clean['annual_salary'] = jobs_df_clean.apply(make_annual, axis=1)
```
